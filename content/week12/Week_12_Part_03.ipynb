# Week 12 Part 3 #

This chapter is about predicting a random variable $Y$ based on a related random variable $X$. There is no restriction on what kind of function of $X$ you are allowed to use as your predictor (apart from our usual assumption that all expectations we are discussing exist).

For a random pair $(X, Y)$, and natural predictor of $Y$ based on $X$ is $E(Y \mid X)$. Recall some [properties of conditional expectation](http://prob140.org/textbook/Chapter_09/02_Expectation_by_Conditioning.html#Other-Properties-of-Conditional-Expectation):

When you condition on $X$, functions of $X$ can be treated as constants. For example:

- $E(g(X) \mid X) = g(X)$
- $E(Yg(X) \mid X) = g(X)E(Y \mid X)$

Keep in mind that these are equalities of random variables: the quantities on both sides of the equations are functions of $X$.

## Reading: Error is Uncorrelated With Functions of $X$ ##

The key observation is that using $E(Y \mid X)$ to predict $Y$ is the equivalent of dropping a perpendicular from $Y$ to the space of all functions of $X$. In random variable language, the error in the prediction is [uncorrelated](http://prob140.org/textbook/Chapter_22/01_Conditional_Expectation_Projection.html#$D_w$-and-Functions-of-$X$) with functions of $X$.
